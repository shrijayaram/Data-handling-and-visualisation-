```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Date":pd.to_datetime(["2022-01-01","2022-03-01","2022-04-01","2022-06-01",
                           "2023-07-01","2023-09-01","2023-10-01","2024-11-1","2024-12-1"]),
    "Value":[10,15,22,19,40,48,42,50,39]
}
```

```python
df = pd.DataFrame(data)
df.set_index("Date", inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"], marker='x')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with Trend")
plt.grid(True)
plt.show()
```

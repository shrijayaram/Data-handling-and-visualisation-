```python
import folium
mymap = folium.Map(location = [12.9166, 77.6101],
                   width = 950,
                   height =550,
                   zoom_start = 12,
                   tiles = 'openstreetmap')

marker = folium.Marker(
    location=[12.9166, 77.6101],
    popup='BTM'
)
mymap.add_child(marker)
mymap

mymap = folium.Map(location=[12.9166, 77.6101], zoom_start=13)

html = """
<img src="https://i0.wp.com/localfeedback.org/wp-content/uploads/2018/03/residential-locality-btm-2nd-stage-bangalore-rubi-ebnath.jpg?fit=1160%2C652&ssl=1" width="200px">
"""

marker = folium.Marker(
    location=[12.9166, 77.6101],
    popup=folium.Popup(html=html)
)
mymap.add_child(marker)
mymap

mymap = folium.Map(location = [13.1155, 77.6070],
                   width = 950,
                   height =550,
                   zoom_start = 12,
                   tiles = 'openstreetmap')

marker = folium.Marker(
    location=[13.1155, 77.6070],
    popup='yelahanka'
)
mymap.add_child(marker)
mymap

mymap = folium.Map(location=[13.1155, 77.6070], zoom_start=13)

html = """
<img src="https://content.jdmagicbox.com/comp/bangalore/z7/080pxx80.xx80.180223205120.z4z7/catalogue/yelahanka-5th-phase-park-yelahanka-new-town-bangalore-parks-mhp9x9q5ux-250.jpg">
"""

marker = folium.Marker(
    location=[13.1155, 77.6070],
    popup=folium.Popup(html=html)
)
mymap.add_child(marker)
mymap
```
